```python
# 누적수익률 구하기 

누적수익률 = 1

for year in range(2000, 2021):
    buy_mon = str(year) + '-11'
    sell_mon = str(year+1) + '-04'
    #print(buy_mon, sell_mon)

    # 매수가
    매수가 = kospi.loc[buy_mon].iloc[0]['Open']
    #print(매수가) # 2000년 첫번째 거래일 524.69
# 매수 월 buy_mon의 첫 번째 거래일 iloc[0] 
    # 매도가
    매도가 = kospi.loc[sell_mon].iloc[-1]['Close']
    #print(매도가) 2001년 마지막 거래일 #577.36
# 매도 월 sell_mon 의 마지막 거래일 iloc[-1]
    # 수익률 
    수익률 = 매도가/매수가 # 577.36 / 524.69 = 1.100383083344451
    
    누적수익률 = 누적수익률 * 수익률 # 1 * 1.100383083344451 

print(누적수익률)
```

```python
#연 평균 수익률 CAGR 구하기 
# 기하 평균 공식 마지막 값 / 처음값 ^ (1/투자기간) - 1

# 누적수익률의 매도가/매수가 가 마지막값/최초값
# ^ = ** 
# 투자기간이 21년이므로 (1/21) -1
# 최종 (누적수익률 ** (1/21)) - 1

CAGR = (누적수익률 ** (1/21)) - 1 
# 투자기간 2000~2021년 까지 21년 이므로 
CAGR * 100
```

```python
# 할로윈 투자전략(11월 매수 4월 매도)를 사용하지 않고 쭉 보유만 했을 때 수익률  

단순보유누적수익률 = kospi.iloc[-1]['Close'] / kospi.iloc[0]['Open'] #매도거-매수가
단순보유CAGR = (단순보유누적수익률 ** (1/21)) -1
단순보유CAGR * 100
```

# 할로윈 투자전략 백테스팅
우리는 할로윈투자전략(11월 매수 4월 매도)이 그냥 가지고만 있었을 때 보다 수익률이 2배가량 높은 것을 확인 했다.

이번에는 매수월을 변경해가면서 정말로 매년 11월에 사서 6개월 동안 투자했을 때 가장 수익이 높았는지를 백테스팅 해보겠다.

```python
#투자 시작일부터 6개월 후를 계산하는 코드

import datetime
from dateutil.relativedelta import relativedelta

start= datetime.datetime(year=2000, month=11, day=1) #시작 2000년 11월 1일
end = start + relativedelta(months=5) # 끝 5달뒤 month=5 2001년 4월 기준날짜에 대한 연산 할 때 쓰는 relativedlta 함수 

# print(start)
# print(end)
# 2000-11-01 00:00:00
# 2001-04-01 00:00:00


#문자열료 년 월 까지만 출력 strftime

# print(start.strftime("%Y-%m"))
# print(end.strftime("%Y-%m"))
```

```python
def 투자6개월(df, start_year=2000, end_year=2020, month=11): #함수 정의 데이터프레임,시작년도,종료년도,매수월 입력 인자 받기
    
    for year in range(start_year, end_year):
        start = datetime.datetime(year=year, month=month, day=1)
        end = start + relativedelta(months=5)
        
        print(start.strftime("%Y-%m"), end.strftime("%Y-%m"))
        
        
투자6개월(None, start_year=2000, end_year=2020, month=12) 
#데이터베이스를 None으로 놓고 시작년 2000 끝나는 년도 2020 매수월을 12월로 입력 
```

```python
#시작월에 매수하고 종료월에 매도하여 수익률을 계산하는 코드

def 투자6개월(df, start_year=2000, end_year=2020, month=11):
    누적수익률 = 1 #클래스 변수 초기화
    
    for year in range(start_year, end_year):
        start = datetime.datetime(year=year, month=month, day=1)
        end = start + relativedelta(months=5)
        
        buy_mon = start.strftime("%Y-%m")
        sell_mon = end.strftime("%Y-%m")
        
        매수가 = df.loc[buy_mon].iloc[0]['Open']
        매도가 = df.loc[sell_mon].iloc[-1]['Close']
        
        수익률 = 매도가/매수가
        누적수익률 = 누적수익률 * 수익률
    return 누적수익률
```

```python
투자6개월(kospi, start_year=2000, end_year=2021, month=11)
```

```python
for month in range(1, 12+1): #1월부터 12월까지 반복문 
    ret = 투자6개월(kospi, start_year=2000, end_year=2021, month=month) #투자6개월 함수를 ret에 바인딩
    #month=month 이므로 1월부터 12월 까지 범위임 
    print(f"{month:02} {ret:.2f}") # 2f 소수점 둘 째 짜리까지 ex) 1.64 1.72 
    
    
    #month를 01로 두면 1 2 3 4 02로 두면 01 02 03 04 03으로 두면 001 002 003 004 처럼 출력  
    
    
    # 21년동안 11월에 매수해서 6개월 보유후 팔았을 때가  누적수익률 6.19로 가장 높다.
```

# 할로윈 투자 전략 시각화

```python
data = {} 

for month in range(1,12+1):
    ret = 투자6개월(kospi, start_year=2000, end_year=2021, month=month)
    data[month] = ret #1월 부터 12월 까지의 월과 수익률을 딕셔너리에 바인딩 
    
print(data)    

#key 값은 투자 시작 월 value 값은 21년간 누적수익률
```
