# get_market_cap 시가총액 조회

```python
df = stock.get_market_cap("20190101")
df.head()
```

```python
#2000년도와 2019년도의 시가총액 top5 비교 


df = stock.get_market_cap("20000104") #2000년 시가총액 조회
s0 = df['시가총액'].sort_values(ascending=False).iloc[:5] #'시가총액' 컬럼 상위 5개값(0,1,2,3,4) 데이터를 s0에 바인딩

df = stock.get_market_cap("20190102") #2019년 시가총액 조회 
s1 = df['시가총액'].sort_values(ascending=False).iloc[:5] 

data = [s0.reset_index(), s1.reset_index()] #reset_index 인덱스를 처음부터 재배열 해주는 함수 s0과 s1을 재배열 해서 data에 바인딩
df = pd.concat(data, keys=["2000년도","2019년도"], axis=1) #concat 연결하는 함수
# s0과 s1 데이터를 바인딩한 data를 데이터 keys 값 2000년도 2019년도 (위에 뜸)
df.head()


# axis = 0 행방향 작업결과가 행으로 나타남 가로
# axis = 1 열방향 작업결과가 열로 나타남 세로
```
