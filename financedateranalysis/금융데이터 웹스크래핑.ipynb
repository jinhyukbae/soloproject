```python
t0 = stock.get_market_ticker_list("19950502")
t1 = stock.get_market_ticker_list("20221001")

print(len(t0), len(t1)) #각날짜의 티커 개수 출력

intersection = set(t0) & set(t1) #집합형 자료구조 set 활용 교집합 &
print(len(intersection)) #t0과 t1에 공통으로 존재하는 티커를 출력

complement = set(t1) - set(t0) #차집합
print(len(complement)) #t0에 없고 t1에만 존재하는 티커 출력 

#intersection 시작일부터 종료일까지 상장유지중인 종목
#complement 새롭게 상장된 종목 
```

```python
# 짧은 시간에 많은 데이터를 요청하면 krx 서버에서 차단 당할 수 있으므로 time 모듈로 쉬면서 데이터를 가져와야 함 
import time
tickers = stock.get_market_ticker_list()
for t in tickers:
    df = stock.get_market_ohlcv("20210101","20210131", t) #날짜의 전종목 ohlcv 데이터를 df에 바인딩
    df.to_excel(f"{t}.xlsx") #데이터를 t라는 이름의 엑셀파일로 저장
    time.sleep(1) # 쉬는 시간
    
   
    
    
```
